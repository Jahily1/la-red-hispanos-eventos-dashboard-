```python
import pandas as pd
asistencia=pd.read_csv('Check-ins eventos_10-08-2025.csv')
asistencia.head()
```

```python
#The first row is deleted from the dataset
asistencia = asistencia.drop(asistencia.index[0])
asistencia.head()
```

```python
#The columns are converted in rows for future analysis
asistencia_larga=asistencia.melt(
    id_vars=['Member Name'],          
    var_name='Evento',            
    value_name='Asistencia'
)

print("\nDataFrame after melt:")
asistencia_larga.head()
```

```python
# The column 'Evento' contains the name of the event and the date
# In the folowing lines, the column is splited between Evento and Fecha
split_name_date = asistencia_larga['Evento'].str.split('(', n=1, expand=True)
asistencia_larga['Fecha']=split_name_date[1]

#The column 'Fecha' is cleaned, This removes everything that is NOT a digit or a dash from anywhere in the string
asistencia_larga['Fecha'] = asistencia_larga['Fecha'].str.replace(r'[^\d\-]', '', regex=True)

# Display the updated DataFrame
print("\nDataFrame after splitting:")
print(asistencia_larga)
```

```python
# Create a temporary series to test the conversion
tested_dates = pd.to_datetime(asistencia_larga['Fecha'], errors='coerce')

# Filter the original table to see ONLY where the conversion failed
errors = asistencia_larga[tested_dates.isna()]

# Show the unique values that are causing the problem
print("Values that do not have the correct format:")
print(errors['Fecha'].unique())
```

```python
# Shows the current data type to see if it's necessary a future type transformation 
asistencia_larga.info()
```

```python
# Data Type modification
# Convert attendance to numbers (for future calcules)
asistencia_larga['Asistencia'] = asistencia_larga['Asistencia'].astype(int)

# Convert date to real time (for future analysis)
asistencia_larga['Fecha'] = pd.to_datetime(asistencia_larga['Fecha'])

# verification type modification
print(asistencia_larga.info())
```

```python
# Create a table that summarize per event, how many people attented
Event_Attendance=asistencia_larga.pivot_table(index='Evento',values='Asistencia', aggfunc='sum')
Event_Attendance.sort_values(by='Asistencia', ascending=False, inplace=True)
Event_Attendance
```

```python
# Importing second database from Evenbrite

# List to keep columns I just want to import
eventbrite_df = pd.read_csv('Attendees_EvenBrite.csv', encoding='mac_roman')
cols_to_keep = ['Attendee first name','Attendee last name','Event name','Event start date']  
eventbrite_df = eventbrite_df[cols_to_keep]
eventbrite_df.head()
```

```python
# Convert date to real time (for future analysis)
eventbrite_df['Event start date'] = pd.to_datetime(eventbrite_df['Event start date'])
eventbrite_df.info()
```

```python
# See what event names you have in each dataset (Heylo and Evenbrite)
print("=== Eventbrite ===")
print(eventbrite_df['Event name'].unique())

print("\n=== Heylo ===")
print(asistencia_larga['Evento'].unique())
```

```python
# Clean both columns for comparison, dropping NaN values
eb_names = set(
    eventbrite_df['Event name']
    .dropna()
    .str.replace(r'\s*\(.*?\)', '', regex=True)
    .str.strip()
    .unique()
)

heylo_names = set(
    asistencia_larga['Evento']
    .dropna()
    .unique()
)

# Events that match exactly
exact_matches = eb_names & heylo_names
print("=== EXACT MATCHES ===")
for e in sorted(exact_matches):
    print(f"  ✓ {e}")

# Events in Eventbrite that DON'T match Heylo
print("\n=== EVENTBRITE ONLY (no match) ===")
for e in sorted(eb_names - heylo_names):
    print(f"  ? {e}")
```
